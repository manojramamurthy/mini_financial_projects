# Part 4: Basic Financial Analysis




## Daily Percentage Change
First we will begin by calculating the daily percentage change. Daily percentage change is defined by the following formula:

$ r_t = \frac{p_t}{p_{t-1}} -1$

2 ways to calculate returns:
(1) a simple calculation using the .shift() method that follows the formula above 
(2) pandas' built in pct_change method. 

Let's create a new column for each dataframe called returns. This column will be calculated from the Close price column.

```python
# Method 1: Using shift
delta['returns'] = (delta['Close'] / delta['Close'].shift(1) ) - 1
```

```python
delta.head()
```

```python
delta['returns'] = delta['Close'].pct_change(1)
```

```python
delta.head()
```

```python
# Now repeat for the other dataframes
JB['returns'] = JB['Close'].pct_change(1)
SW['returns'] = SW['Close'].pct_change(1)
```

```python
JB.head()
```

```python
SW.head()
```

** Create a cumulative daily return column for each airline company's dataframe.**

```python
delta['Cumulative Return'] = (1 + delta['returns']).cumprod()
```

```python
delta.head()
```

```python
JB['Cumulative Return'] = (1 + JB['returns']).cumprod()
SW['Cumulative Return'] = (1 + SW['returns']).cumprod()
```

** Now plot the Cumulative Return columns against the time series index. Which stock showed the highest return for a $1 invested? Which showed the lowest?**

```python
delta['Cumulative Return'].plot(label='DAL',figsize=(16,8),title='Cumulative Return')
JB['Cumulative Return'].plot(label='JBLU')
SW['Cumulative Return'].plot(label='LVU')
plt.legend();
```
